```python
def get_main_urls(url):
    with open('main_urls.txt', 'w') as out:
        driver = webdriver.Chrome(executable_path='/home/artem/ChromeDriver/chromedriver')
        driver.get(url)
        sleep(1)
        counter = 1
        acces_exist = True
        while acces_exist == True:
            try:
                acces = driver.find_element(By.XPATH, f'/html/body/div/div/mibig-repository/div[2]/table/tbody/tr[{counter}]')
                for i in acces.find_elements(By.CLASS_NAME, 'ng-binding'):
                    out.write(i.text+' | ')
                out.write('\n')
                counter += 1
            except Exception as ex:
                print(ex)
                acces_exist = False
```

```python
def get_fasta(url_code):
    with open(os.getcwd()+'/seq/'+url_code+'.txt', 'w') as file:
        url = 'https://mibig.secondarymetabolites.org/repository/'+url_code
        driver = webdriver.Chrome(executable_path='/home/artem/ChromeDriver/chromedriver')
        driver.get(url)
        driver.find_element(By.XPATH, '//*[@id="r1c1"]/div/div[1]/div[1]/div[4]/a').click()
        if len(driver.find_elements(By.CLASS_NAME, 'hi_warn')) == 1:
            print('')
            driver.find_element(By.XPATH, '//*[@id="maincontent"]/div/div[5]/div[1]/ul/li/span/a').click()
        
        for element in driver.find_elements(By.CLASS_NAME, 'brieflinkpopperctrl'):
            if element.text == 'Taxonomy':
                element.click()
                break
        for element in driver.find_elements(By.CLASS_NAME, 'brieflinkpopperctrl'):
            if element.text == 'Nucleotide':
                element.click()
                break
        driver.find_element(By.XPATH, '//*[@id="maincontent"]/div/div[5]/div[1]/div[2]/p/a').click()
        if 'contains no sequence data' in driver.find_element(By.CLASS_NAME, 'seqviewer').text:
            file.write('This entry is the master record for a whole genome shotgun sequencing project and contains no sequence data.\n')
        else:
            for element in driver.find_elements(By.CLASS_NAME, 'dblinks'):
                if element.text == 'FASTA':
                    element.click()
                    break
        sleep(2)
        while len(driver.find_elements(By.CLASS_NAME, 'loadingbar_update_text')) != 0:
             sleep(5)
        content = driver.find_element(By.ID, 'viewercontent1').text
        file.write(content)
        print('Done for ' + url_code)
```
