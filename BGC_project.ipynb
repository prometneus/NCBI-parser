```python
def urls_generator(upper_bound):
    for i in range(1, upper_bound):
        yield str('/home/sirius/BGCproject/mibig_json_2.0/BGC000' + (4-len(str(i)))*'0' + str(i) + '.json')
```

```python
def names_definition(n):
    counter = 0
    names = []
    types = []
    for url in urls_generator(n):
        try:
            with open(url) as f:
                data = json.load(f)
                for elem in data['cluster'].items():
                    if 'organism_name' in elem:
                        names.append(elem[1])
                    if 'biosyn_class' in elem:
                        types.append([elem[1]])
        except:
            counter += 1
            f.close()
    print(f'{counter} urls didn\'t exist')
    f.close()
    for i in range(len(names)):
        names[i] = [names[i], types[i]]
    return names
```

```python
def ncbi_parsing(name):
    complete_genomes, chromosomes = [], []
    scaffolds, contigs = [], []
    for line in fdf:
        try:     # 7 -- organism name, 11 -- assembly level
            if name in line[7]:
                if re.search('Complete Genome', line[11]) and re.search('ftp', line[19]):
                    complete_genomes.append([name, 'Complete Genome', line[19] + '/' + line[19].split('/')[-1] + '_genomic.fna.gz']) #link in line[19]
                elif re.search('Chromosome', line[11]) and re.search('ftp', line[19]): #'Chromosome' in line[11]:
                    chromosomes.append([name, 'Chromosome', line[19] + '/' + line[19].split('/')[-1] + '_genomic.fna.gz'])
                elif re.search('Scaffold', line[11]) and re.search('ftp', line[19]): #'Scaffold' in line[11]:
                    scaffolds.append([name, 'Scaffold', line[19] + '/' + line[19].split('/')[-1] + '_genomic.fna.gz'])
                elif re.search('Contig', line[11]) and re.search('ftp', line[19]):
                    contigs.append([name, 'Contig', line[19] + '/' + line[19].split('/')[-1] + '_genomic.fna.gz'])
        except:
            pass
            #print(f'Problem with a {name} occured')
    if len(complete_genomes) != 0: 
        return complete_genomes
    elif len(chromosomes) != 0: 
        return chromosomes
    elif len(scaffolds) != 0: 
        return scaffolds
    elif len(contigs) != 0:
        return contigs
```

```python
start_time = time.time()
official_names = names_definition(n=2040)
clean_names = []
for data in official_names:
    if data not in clean_names:
        clean_names.append(data)
    else:
        for line in clean_names:
            if data in line:
                clean_names[1].append(data[1])

array = []
for appelation in clean_names:
    array.append(ncbi_parsing(name=appelation[0]))
clean_array = []
for elem in filter(None. __ne__, array):
    clean_array.append(elem)
for name in clean_names:
    for element in clean_array:
        if re.search(name[0], element[0][0]) and element[0][2] not in name:
            name.append(element[0][1])
            name.append(element[0][2])

final_table = []
for element in clean_names:
    if len(element) > 2:
        final_table.append(element)

for line in final_table:
    for data in official_names:
        if data[0].strip() == line[0].strip():
        #if re.search(data[0], line[0]):
            line[1].append(data[1].copy())
    line[1].pop(0)
    
print(len(final_table))
print(f'Time spent: {time.time() - start_time}')
```
