```python
import json
import pandas as pd
import numpy as np
import requests
import time
import csv
import re
```

```python
def urls_generator(upper_bound):
    for i in range(1, upper_bound):
        yield str('/home/sirius/BGCproject/mibig_json_2.0/BGC000' + (4-len(str(i)))*'0' + str(i) + '.json')
```

```python
def names_definition(n):
    counter = 0
    names = set()
    for url in urls_generator(n):
        try:
            with open(url) as f:
                data = json.load(f)
                for elem in data['cluster'].items():
                    if 'organism_name' in elem:
                        names.add(elem[1])
                        f.close()
        except:
            counter += 1
            #print(f'{url} doesn\'t exist')
            f.close()
    print(f'{counter} urls didn\'t exist')
    #print(names)

    return names
```

```python
def ncbi_parsing(name):
    complete_genomes, chromosomes = [], []
    scaffolds, contigs = [], []
    for line in fdf:
        try:     # 7 -- organism name, 11 -- assembly level
            if name in line[7]:
                if re.search('Complete Genome', line[11]):
                    complete_genomes.append(line)
                elif re.search('Chromosome', line[11]): #'Chromosome' in line[11]:
                    chromosomes.append(line)
                elif re.search('Scaffold', line[11]): #'Scaffold' in line[11]:
                    scaffolds.append(line)
                elif re.search('Contig', line[11]):
                    contigs.append(line)
        except:
            print(f'Problem with a {name} occured')
            NameError
    if len(complete_genomes) != 0: 
        return complete_genomes
    elif len(chromosomes) != 0: 
        return chromosomes
    elif len(scaffolds) != 0: 
        return scaffolds
    elif len(contigs) != 0:
        return contigs
    else:
        print(f'No full sequence for {name} was found')
```

```python
start_time = time.time()
official_names = names_definition(n=2040)
final_array = []
for appelation in official_names:
    final_array.append(ncbi_parsing(name=appelation))
#print('\n\n\nFinal Array looks like: ', final_array)
print(len(final_array))
print(f'Time spent on parsing {time.time() - start_time}s')
```
